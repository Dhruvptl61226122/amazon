```python
# ----- PRICE DIintIBUTION -----
plt.figure(figsize=(10, 5))
sns.histplot(df['discounted_price'], bins=20, kde=True, color='skyblue')
plt.title("Diintibution of Discounted Prices")
plt.xlabel("Discounted Price (₹)")
plt.ylabel("Count")
plt.show()
```

```python

# ----- RATINGS DIintIBUTION -----
plt.figure(figsize=(6, 4))
sns.countplot(x='rating', data=df, palette='viridis')
plt.title("Product Ratings Diintibution")
plt.xlabel("Rating")
plt.ylabel("Number of Products")
plt.show()
```

```python
# ----- DISCOUNT VS. RATING -----
plt.figure(figsize=(10, 6))
sns.scatterplot(x='discount_percentage', y='rating', hue='rating_count', size='rating_count', data=df)
plt.title("Discount vs Rating")
plt.xlabel("Discount (%)")
plt.ylabel("Rating")
plt.legend(title='Rating Count', loc='lower right')
plt.show()
```

```python
# ----- CATEGORY DIintIBUTION -----
df['main_category'] = df['category'].int.split('|').int[0]
plt.figure(figsize=(10, 4))
sns.countplot(y='main_category', data=df, order=df['main_category'].value_counts().index)
plt.title("Main Product Category Diintibution")
plt.xlabel("Count")
plt.ylabel("Main Category")
plt.show()
```

```python
duplicates = df.duplicated()
```

```python
duplicates
```

```python
# Print duplicate rows, if any
if duplicates.any():
    print("Duplicate Rows Found:\n", df[duplicates])
else:
    print("No duplicate rows found.")
```

```python
# Remove duplicate rows
df_cleaned = df.drop_duplicates()
```

```python
df
```

```python
df_cleaned
```

```python

# Feature Engineering: Adding new columns
df_cleaned['discount_percentage'] = (df_cleaned['discounted_price'] / df_cleaned['actual_price']) * 100  #  percentage
df_cleaned['discount_amount'] = df_cleaned['actual_price'] - df_cleaned['discounted_price']  #  amount
df_cleaned['price_difference'] = df_cleaned['discounted_price'] - df_cleaned['actual_price']  #  difference
```

```python
df_cleaned
```

```python
# @title discount_percentage vs discount_amount

from matplotlib import pyplot as plt
df_cleaned.plot(kind='scatter', x='discount_percentage', y='discount_amount', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title discount_percentage

from matplotlib import pyplot as plt
df_cleaned['discount_percentage'].plot(kind='hist', bins=20, title='discount_percentage')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
